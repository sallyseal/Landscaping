```python
# This script replaces missing values with 0s in the combined data file
# This script reads in the GOI file and writes the GOI to a list
```

```python
import pandas as pd
import numpy as np
```

```python
# Read in the combined data file that was made in Excel by combining the 3 cleaned datasets

df = pd.read_csv('combined_data.csv')
df.head()
```

```python
# Fill all missing values with 0 as Julia cannot handle dataframes with missing values

df1 = df.fillna(0)
df1.head()
df1.to_csv('combined_data.csv')
```

```python
# Get the GOIs in a list

data = []
with open('Blood_GOI.csv') as infile:
    for line in infile.readlines():
        temp = line.strip().split(',')[1]
        data.append(str(temp))

data.pop(0)
GOI_list = data
print(GOI_list)
```

```python
# Iterate through the df and extract only the information of the genes present in the GOI_list

GOI_data = []
for i, row in df1.iterrows():
    for j in GOI_list:
        if row[1] == j:
            GOI_data.append(row)
    print(GOI_data)
```

```python
# Put GOI_data back into a df and write it to a file
# From here we can extract arrays of info for each of the 26 genes and feed into mutual information function in Julia

GOI_df = pd.DataFrame(GOI_data)
GOI_df.head()
GOI_df.to_csv('GOI.csv')
```

```python
# Generate all possible combinations of the 26 genes for MI calculations
# n choose k, where n = 26 and k = 2 as it's pairwise comparison

import itertools as it

combinations =  (list(it.combinations(['CD44', 'CSF3R', 'CD135/FLT3', 'SOCS3', 'CD123/IL3RA', 'CD42/GP1BA', 'CD41/ITGA2B', 'GATA1', 'GATA2', 'CNRIP1', 'KLF1', 'LEF1', 'CD71/TFRC', 'TMOD1', 'ANK1', 'CD235a/GYPA', 'MYB', 'VWF', 'FLI1', 'NFIB', 'TGFB1', 'TGFB2', 'LOX', 'CD61/ITGB3', 'CLU'], 2)))
print(combinations)
print(len(combinations))

# 325 combinations are possible when n = 26 and k = 2
```

```python
# Generate an array for each of the 26 GOI to feed into the MI function in Julia. Julia array = Python list
# E.g. Gene_id = [] etc.

GOI_array = GOI_df.values.tolist()
print(GOI_array[0])
print(GOI_array[1])
print(GOI_array[24])
# Then switch to Julia and compute the 325 MI's and select the pairs with highest MI

# Can also consider doing all possible pairs of combinations and calculating the MI and then see if any of those
# unique paris have a higher mutual information than any of the 325 calculated pairs
   
```

```python
# Make a dictionary where each gene is the key and the value is a list of all the values

# First iterate through the GOI_array and append each GOI Gene ID to a list and associated values to another list
GOI_pre_dict = []
GOI_values = []
for i in GOI_array:
    GOI_pre_dict.append(i[1])
    GOI_values.append(i[2:])
print(GOI_pre_dict)
print(GOI_values)
```

```python
# Now make a dictionary from these two lists and we will use this to feed into MI 

GOI_dict = dict(zip(GOI_pre_dict, GOI_values))
print(GOI_dict)
```

```python
# Iterate through the dataframe and replace all 0 values with 1
# Iterate through the dataframe and replace all values != 1 with 0
# Plot this to see the distribution of zeros in the data and see if there are any trends
import pandas as pd

df = pd.read_csv('combined_data_no_header.csv', header=None)
df.head()

df1 = df.replace(0, 1)
df1.head()

df2 = df1.apply(lambda x: [y if y == 1 else 0 for y in x])
df2.head()
df2.to_csv("combined_data_1_0.csv")
```
